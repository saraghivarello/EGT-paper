```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
grad_opinion=LinearSegmentedColormap.from_list('my_gradient', (
    # Edit this gradient at https://eltos.github.io/gradient/#CC0000-FFBA0A-00523A
    (0.000, (0.800, 0.000, 0.000)),
    (0.500, (1.000, 0.729, 0.039)),
    (1.000, (0.000, 0.322, 0.227))))
grad_coop= LinearSegmentedColormap.from_list('my_gradient', (
    # Edit this gradient at https://eltos.github.io/gradient/#D90079-FDFFFC-003C7E
    (0.000, (0.851, 0.000, 0.475)),
    (0.500, (0.992, 1.000, 0.988)),
    (1.000, (0.000, 0.235, 0.494))))
```

```python
# Cooperation fraction from area

wordset2 = np.genfromtxt(fname='coop41.tsv')
wordset2_r = wordset2[::-1]
df2 = pd.DataFrame(wordset2_r,index=np.linspace(0,1,len(wordset2_r))[::-1].round(decimals=3),
                   columns=np.linspace(0,1,len(wordset2_r)).round(decimals=3))
fig,ax = plt.figure(figsize=(10, 8), dpi=80), plt.axes()
p2 = sns.heatmap(df2,cmap=grad_coop,cbar=False,annot=False,vmin=0,vmax=1)
cbar = plt.colorbar(p2.collections[0])
cbar.set_label(label='Final cooperation $\mathregular{C_f}$',size=35),cbar.set_ticks([0, 0.5, 1]),cbar.ax.tick_params(labelsize=20) 
ax.set_xticks(np.linspace(0,41,5)),ax.set_xticklabels(np.linspace(0,1,5).round(decimals=2),size=20),ax.set_yticks(np.linspace(0,41,5)),ax.set_yticklabels(np.linspace(0,1,5)[::-1].round(decimals=2),size=20)
plt.xlabel('$\mathregular{p_0}$',fontsize=35),plt.ylabel('$\mathregular{c_n}$',fontsize=35)
x = np.linspace(0,len(wordset2),10)
ax.plot(x, len(wordset2)-x*0.9, color='k', lw=4),ax.plot(x, len(wordset2)-x*0.9*0.75, color='#FFBA0A', lw=4)
ax.plot(x, len(wordset2)-x*0.9*0.5, color='w', lw=4)
ax.plot(41*0.7,len(wordset2)-0.25*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.plot(41*0.7,len(wordset2)-0.5*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.plot(41*0.4,len(wordset2)-0.25*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.plot(41*0.4,len(wordset2)-0.5*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.plot(41*0.1,len(wordset2)-0.25*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.plot(41*0.1,len(wordset2)-0.5*41, marker="o", markersize=12, markeredgecolor="#003C7E", markerfacecolor="w")
ax.annotate('f', (41*0.7,len(wordset2)-0.25*41), fontsize=20,xytext=(41*0.7+1, len(wordset2)-0.25*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
ax.annotate('c', (41*0.7,len(wordset2)-0.5*41), fontsize=20,xytext=(41*0.7+1, len(wordset2)-0.5*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
ax.annotate('e', (41*0.4,len(wordset2)-0.25*41), fontsize=20,xytext=(41*0.4+1, len(wordset2)-0.25*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
ax.annotate('b', (41*0.4,len(wordset2)-0.5*41), fontsize=20,xytext=(41*0.4+1, len(wordset2)-0.5*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
ax.annotate('d', (41*0.1,len(wordset2)-0.25*41), fontsize=20,xytext=(41*0.1+1, len(wordset2)-0.25*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
ax.annotate('a', (41*0.1,len(wordset2)-0.5*41), fontsize=20,xytext=(41*0.1+1, len(wordset2)-0.5*41-2),color='k',bbox=dict(facecolor='w', edgecolor='black', boxstyle='round'))
plt.show()
```
