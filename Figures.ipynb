```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
folder_path = r'C:\Users\PcLaptop\Desktop\EGT paper'
files = os.listdir(folder_path)         # Get a list of all files in the folder

for file_name in files:
    file_path = os.path.join(folder_path, file_name)  # Construct the file path

    if os.path.isfile(file_path) and file_name.endswith('.txt'):       # Check if the item in the folder is a file
        with open(file_path, 'r') as file:   # Open the file for reading      
            content = file.readlines()     # Read the contents of the file
       
        var_name = os.path.splitext(file_name)[0]     # Create a variable with the same name as the file (without the extension)
        array = []

        for line in content: # Iterate over the lines in the file           
            line = line.strip()  # Remove any trailing whitespace or newline characters
            array.append(float(line))   # Convert the line to a floating-point number and append it to the array

        locals()[var_name] = array  # Assign the array to the variable with the same name as the file
```
