```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import matplotlib.gridspec as gridspec
```

```python
folder_path = r"C:\Users\user\Documents\GitHub\EGT-paper\Fixation probability\FP files"
files = os.listdir(folder_path)         # Get a list of all files in the folder

for file_name in files:
    file_path = os.path.join(folder_path, file_name)  # Construct the file path

    if os.path.isfile(file_path) and file_name.endswith('.txt'):       # Check if the item in the folder is a file
        with open(file_path, 'r') as file:   # Open the file for reading      
            content = file.readlines()     # Read the contents of the file
       
        var_name = os.path.splitext(file_name)[0]     # Create a variable with the same name as the file (without the extension)
        array = []

        for line in content: # Iterate over the lines in the file           
            line = line.strip()  # Remove any trailing whitespace or newline characters
            array.append(float(line))   # Convert the line to a floating-point number and append it to the array

        locals()[var_name] = array  # Assign the array to the variable with the same name as the file
```

```python
# Data placeholders (replace with actual data)
f_data = [
    [f4251_5_0, f42501_5_0, f425001_5_0],
    [f451_5_0, f4501_5_0, f45001_5_0],
    [f4751_5_0, f47501_5_0, f475001_5_0]
]
colors = ['moccasin', 'lightskyblue', 'slateblue']
annotations = [
    (15, 0.2, r'$c_n$ = 0.25' + '\n' + 'Harmony\n Game'),
    (4, 0.7, r'$c_n$ = 0.5' + '\n' + 'Snow Drift'),
    (4, 0.6, r'$c_n$ = 0.75' + '\n' + "Prisoner's\n Dilemma")
]

# Figure setup
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(4, 7))
gs = gridspec.GridSpec(3, 1)
gs.update(hspace=0)

xmin, xmax = -1, 21
x = np.linspace(0, 20, 1000)
y = 1 / 20 * x

# Plot each subplot
for i, ax in enumerate(axs):
    ax.plot(np.arange(21), f_data[i][0], marker='o', linewidth=2, label=r'$\beta$ = 1')
    ax.plot(np.arange(21), f_data[i][1], marker='o', linewidth=2, label=r'$\beta$ = 0.1')
    ax.plot(np.arange(21), f_data[i][2], marker='o', linewidth=2, label=r'$\beta$ = 0.01')
    ax.axvspan(xmin, xmax, facecolor=colors[i], alpha=0.4)
    ax.text(*annotations[i], fontsize=12, ha="center")
    ax.set_xlim(right=xmax, left=xmin)
    ax.plot(x, y, '-k', alpha=0.2)
    ax.tick_params(axis='y', labelsize=12)
    if i == 1:
        leg = ax.legend(loc='lower right', fontsize=11)
        leg.get_frame().set_alpha(0)
    if i == 2:
        ax.set_xlabel('Initial number of cooperators', fontsize=12)
        ax.tick_params(axis='x', labelsize=12)
    else:
        ax.set_xticks([])

    ax.set_ylabel('Fixation probability', fontsize=12)

#plt.tight_layout()
# plt.savefig('c_n.pdf')
plt.tight_layout(pad=1.0)
plt.savefig('fixation_probability.pdf')
plt.show()
```
