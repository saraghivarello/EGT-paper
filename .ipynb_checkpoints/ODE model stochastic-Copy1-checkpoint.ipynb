```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def funct_ode(y,t,pars):
    x = y[0]
    a = y[1]
    dxdt = x*(1 - x)*(pars['e']*pars['p']*pars['cd']*(1 - pars['d']*x) - pars['cn']*(1 - a))
    dadt = pars['gamma']*a*(1 - a)*(2*x - 1)
    dydt = np.array([dxdt,dadt])
    return dydt
```

```python
def funct_tan(y,t,pars):
    x = y[0]
    a = y[1]
    dxdt = x*(1 - x)*(np.tanh(pars['beta']/2*(pars['e']*pars['p']*pars['cd']*(1 - pars['d']*x) - pars['cn']*(1 - a)))) + np.sqrt(x*(1-x)/pars['N'])*np.random.normal(0, 1)
    dadt = pars['gamma']*a*(1 - a)*(2*x - 1)
    dydt = np.array([dxdt,dadt])
    return dydt
```

```python
#Stochastic differential equations SDEINT

import numpy as np
import sdeint

A = np.array([[-0.5, -2.0],
              [ 2.0, -1.0]])

B = np.diag([0.5, 0.5]) # diagonal, so independent driving Wiener processes

tspan = np.linspace(0.0, 10.0, 10001)
x0 = np.array([3.0, 3.0])

def f(x, t):
    return A.dot(x)

def G(x, t):
    return B

result = sdeint.itoint(f, G, x0, tspan)

# Visualize
fig1,ax = plt.figure(figsize=(3, 3)), plt.axes()
plt.plot(tspan,result,linewidth=1)
plt.show()
```

```python
# Parameters
pars={} #initiate parameter dictionary
pars['p']=0.9
pars['cd']=1
pars['e']=0.9
pars['d']=0.5
pars['cn']=0.2
pars['gamma']=0.1
pars['beta']=0.1
pars['N']=100000

# Initial conditions
c0 =  0.5   #initial fraction of cooperators 
a0 =  0.5   #initial alpha mean 
y0 = np.array([c0,a0])
t = np.linspace(0,1000,10000)

# Simulate
int_y = integrate.odeint(funct_tan,y0,t,args=(pars,))

# Visualize
fig1,ax = plt.figure(figsize=(7, 7)), plt.axes()
plt.plot(t,int_y[:,0],linewidth=3,color='k')
plt.plot(t,int_y[:,1],linewidth=3,color='r')

plt.xlabel('Time',fontsize=20)
plt.ylabel('Population fraction',fontsize=20)
ax.tick_params(labelsize=15)
#ax.set_xticks(np.linspace(0,,3))
#ax.set_xticklabels(np.linspace(0,100,3).astype(int),fontsize=20) 
ax.set_yticks(np.linspace(0,1,5))
ax.set_yticklabels(np.linspace(0,1,5).round(decimals=2),fontsize=20) 
plt.legend(['x','α'],fontsize=20,loc='lower right')
plt.show()
```

```python
# Parameters
pars={} #initiate parameter dictionary
pars['p']=0.9
pars['cd']=1
pars['e']=0.9
pars['d']=0.5
pars['cn']=0.2
pars['gamma']=1

# Initial conditions
c0 =  0.5   #initial fraction of cooperators 
a0 =  0.5   #initial alpha mean 
y0 = np.array([c0,a0])
t = np.linspace(0,50,10000)

# Simulate
int_y = integrate.odeint(funct_ode,y0,t,args=(pars,))

# Visualize
fig1,ax = plt.figure(figsize=(7, 7)), plt.axes()
plt.plot(t,int_y[:,0],linewidth=3,color='k')
plt.plot(t,int_y[:,1],linewidth=3,color='r')
plt.xlabel('Time',fontsize=20)
plt.ylabel('Population fraction',fontsize=20)
ax.tick_params(labelsize=15)
#ax.set_xticks(np.linspace(0,len(t),3))
#ax.set_xticklabels(np.linspace(0,len(t),3).astype(int),fontsize=20) 
ax.set_yticks(np.linspace(0,1,5))
ax.set_yticklabels(np.linspace(0,1,5).round(decimals=2),fontsize=20) 
plt.legend(['x','α'],fontsize=20,loc='lower right')
plt.show()
```
